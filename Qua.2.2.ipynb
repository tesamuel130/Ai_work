```python
# Cell 3: Implement Uniform Cost Search
def uniform_cost_search(graph, start, goal):
    # Priority queue to store (cumulative_cost, current_node, path)
    pq = PriorityQueue()
    pq.put((0, start, [start]))  # (cost, node, path)

    visited = set()  # To keep track of visited nodes

    while not pq.empty():
        cumulative_cost, current, path = pq.get()

        # If the goal is reached, return the path
        if current == goal:
            return path, cumulative_cost

        # Mark the current node as visited
        visited.add(current)

        # Explore neighbors
        for neighbor, cost in graph.get(current, []):
            if neighbor not in visited:
                new_cost = cumulative_cost + cost
                new_path = path + [neighbor]
                pq.put((new_cost, neighbor, new_path))

    return "No path found."
```

```python
# Cell 4: Test the UCS Algorithm
initial_state = "Addis"
goal_state = "Lallbela"

path, cost = uniform_cost_search(graph, initial_state, goal_state)
print("Path from Addis to Lalibela:", path)
print("Total Cost:", cost)
```
