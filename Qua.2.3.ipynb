```python
# Cell 3: Implement Customized Uniform Cost Search
def multi_goal_uniform_cost_search(graph, start, goals):
    # Priority queue to store (cumulative_cost, current_node, visited_goals, path)
    pq = PriorityQueue()
    pq.put((0, start, [], [start]))  # (cost, node, visited_goals, path)

    remaining_goals = set(goals)  # Goals yet to be visited
    visited = set()  # To keep track of visited nodes globally

    while not pq.empty():
        cumulative_cost, current, visited_goals, path = pq.get()

        # If the current node is a goal and not yet visited, mark it as visited
        if current in remaining_goals and current not in visited_goals:
            visited_goals = visited_goals + [current]  # Append the current goal to visited_goals
            remaining_goals.remove(current)

            # If all goals are visited, return the path and cumulative cost
            if not remaining_goals:
                return path, cumulative_cost

        # Mark the current node as visited
        visited.add(current)

        # Explore neighbors
        for neighbor, cost in graph.get(current, []):
            if neighbor not in visited:
                new_cost = cumulative_cost + cost
                new_path = path + [neighbor]
                pq.put((new_cost, neighbor, visited_goals, new_path))

    # If no path is found, return empty path and infinite cost
    return [], float('inf')
```
